```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import cv2
```

```python
im = cv2.imread('internimage.jpg')
im_resized = cv2.resize(im, (340, 340), interpolation=cv2.INTER_LINEAR)
```

```python

plt.imshow(cv2.cvtColor(im_resized, cv2.COLOR_BGR2RGB))
plt.show()
```

```python

image_grey = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im_resized = cv2.resize(image_grey, (340, 340), interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(im_resized, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image_invert = cv2.bitwise_not(image_grey)
im_resized = cv2.resize(image_invert, (340, 340), interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(im_resized, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
blur = cv2.GaussianBlur(image_invert, (21, 21), 0)
```

```python
blur_invert = 255-blur
```

```python
final_sketch = cv2.divide(image_grey, blur_invert, scale=256.0)
im_resized = cv2.resize(final_sketch, (340, 340), interpolation=cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(im_resized, cv2.COLOR_BGR2RGB))
plt.show()
```
