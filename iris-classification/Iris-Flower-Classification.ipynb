```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
missing_values_count = df.isnull().sum()

total_cells = np.product(df.shape)

total_missing = missing_values_count.sum()

percentage_missing = (total_missing/total_cells)*100
print(percentage_missing)
```

```python

NAN = [(c, df[c].isnull().mean()*100) for c in df]
NAN = pd.DataFrame(NAN, columns=['column_name', 'percentage'])
NAN
```

```python

null_counts = df.isnull().sum()
null_counts
```

```python
for i in df.columns:
    print(df[i].unique())
```

```python
df_setosa = df.loc[df['Species']=='Iris-setosa']
df_virginica = df.loc[df['Species']=='Iris-virginica']
df_versicolor = df.loc[df['Species']=='Iris-versicolor']
```

```python
plt.plot(df_setosa['SepalLengthCm'],np.zeros_like(df_setosa['SepalLengthCm']),'*')
plt.plot(df_virginica['SepalLengthCm'],np.zeros_like(df_virginica['SepalLengthCm']),'o')
plt.plot(df_versicolor['SepalLengthCm'],np.zeros_like(df_versicolor['SepalLengthCm']),'+')
```

```python

plt.plot(df_setosa['SepalWidthCm'],np.zeros_like(df_setosa['SepalWidthCm']),'*')
plt.xlabel('SepalLength')
plt.show()
```

```python
plt.plot(df_virginica['SepalLengthCm'],np.zeros_like(df_virginica['SepalLengthCm']),'o')
plt.xlabel('SepalLength')
plt.show()
```

```python

plt.plot(df_versicolor['PetalLengthCm'],np.zeros_like(df_versicolor['PetalLengthCm']),'+')
plt.xlabel('PetalLengthCm')
plt.show()
```
